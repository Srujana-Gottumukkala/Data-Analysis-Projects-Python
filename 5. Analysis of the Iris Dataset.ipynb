```python
# Boxplot for understanding the data distribution
plt.figure(figsize=(8,6))
sns.boxplot(data=iris)
```

```python
plt.figure(figsize=(6, 5))
custom_colors = ['#0085BB', '#ED7117', '#06AC0C']
sns.countplot(x='Species', data=iris ,palette=custom_colors)
plt.show()
```

```python
sns.lmplot(iris, x='SepalLengthCm', y='PetalLengthCm', height = 6)
plt.show()
```

```python
sns.jointplot(x = 'SepalLengthCm', y='PetalLengthCm',data=iris, kind='hex', gridsize=20, height=6)
```

```python
sns.jointplot(x = 'SepalLengthCm', y='SepalWidthCm',data=iris, height=6)
```

```python
plt.figure(figsize = (6,5))
sns.kdeplot(x='SepalLengthCm', y='SepalWiirdthCm', data=iris)
```

```python
plt.figure(figsize = (6,5))
sns.kdeplot(x='SepalLengthCm', y='SepalWidthCm', data=iris, fill=True, cmap='Blues')
plt.show()
```

```python
fig1 = sns.FacetGrid(iris,col='Species')
fig1.map(sns.histplot,'SepalLengthCm')
plt.show()
```

```python
fig2 = sns.FacetGrid(iris, hue='Species')
fig2.map(plt.scatter, 'SepalLengthCm','SepalWidthCm')
fig2.add_legend()
plt.show()
```
