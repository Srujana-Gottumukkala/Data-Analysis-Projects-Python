### Visualization using Seaborn

```python
# Importing the libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Importing the dataset
df = pd.read_csv(r'C:\Users\sruja\heart.csv')
```

```python
df.head()
```

#### Descriptive Statistics

```python
df.describe()
```

```python
df.dtypes
```

#### Data Analysis of Numerical Data

```python
plt.hist(df['age'],bins = [20,30,40,50,60,70,80])
plt.title('Age')
# 
```

```python
plt.hist(df['trestbps'], bins = [90,100,110,120,130,140,150,160,170,180,190,200])
plt.title('Resting Blood Pressure')
```

```python
plt.hist(df['chol'], bins = 10)
plt.title('Cholesterol')
```

```python
plt.hist(df['thalach'], bins = [60,70,90,100,110,120,130,140,150,160,170,180,190,200], edgecolor = 'black')
plt.title('Max Heart Rate')
```

```python
plt.hist(df['oldpeak'], bins = 10, edgecolor = 'black')
plt.title('ST Depression')
```

```python
plt.scatter(df['age'],df['trestbps'], s=20, c = '#AAAAAA', edgecolor = 'blue', linewidth = 1, alpha = 0.8)
plt.xlabel('Age')
plt.ylabel('Resting Blood Pressure')
plt.title('Age vs RBP')
```

```python
plt.scatter(df['age'],df['chol'], s=20, c = '#BBBBBB', edgecolor = 'lime', linewidth = 1, alpha = 0.8)
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.title('Age vs Cholesterol')
```

```python
plt.scatter(df['age'],df['thalach'], s=20, c = '#CCCCCC', edgecolor = 'teal', linewidth = 1, alpha = 0.8)
plt.xlabel('Age')
plt.ylabel('Max Heart Rate')
plt.title('Age vs Max Heart Rate')
```

```python
plt.scatter(df['age'],df['oldpeak'], s=20, c = '#999999', edgecolor = 'cyan', linewidth = 1, alpha = 0.8)
plt.xlabel('Age')
plt.ylabel('ST depression')
plt.title('Age vs ST depsression')
```

#### Note:
From the above graphs, it can be seen that several parameters such as BP, Cholesterol, etc which are linked to heart disease, do not have any direct relationship with Age

#### Kernal Density Estimates (KDE Plots)

```python
sns.jointplot(x=df['chol'], y=df['trestbps'], data=df, kind="kde")
```

```python
sns.jointplot(x=df['thalach'], y=df['chol'], data=df)
```

```python
f, ax = plt.subplots(figsize=(5, 5))
sns.kdeplot(df['chol'], df['oldpeak'], ax=ax)
sns.rugplot(df['chol'], color="g", ax=ax)
sns.rugplot(df['oldpeak'], vertical=True, ax=ax)
```

```python
plt.scatter(df['trestbps'],df['thalach'], s=20, c = '#EEEEEE', edgecolor = 'blue', linewidth = 1, alpha = 0.9)
plt.xlabel('Resting Blood Pressure')
plt.ylabel('Max Heart Rate')
plt.title('RBP vs Max Heart Rate')
```

```python
plt.scatter(df['trestbps'],df['oldpeak'], s=20, c = '#CCCCCC', edgecolor = 'blue', linewidth = 1, alpha = 0.8)
plt.xlabel('Resting Blood Pressure')
plt.ylabel('ST Depression')
plt.title('RBP vs ST Depression')
```

```python
plt.scatter(df['thalach'],df['oldpeak'], s=40, c = '#FFBB99', edgecolor = 'orange', linewidth = 1, alpha = 1)
plt.xlabel('Max Heart Rate')
plt.ylabel('ST Depression')
plt.title('Max Heart Rate vs ST Depression')
```
