**From the above visualization, it is clear that there are many missing values, with a few columns that consist of far greater missing values than the others. So, we will check the missing values as a percentage of the toal values of each column.**

```python
round(past_applications.isnull().sum()/past_applications.shape[0] * 100,2)
```

Please find below a plot of the missing values with 10% as a cut-off of acceptance percentage.

```python
acceptance_past_applications = pd.DataFrame((past_applications.isnull().sum())*100/past_applications.shape[0]).reset_index()
acceptance_past_applications.columns = ['Column Name', 'Percentage of Null Values']
percent_missing_fig = plt.figure(figsize = (10,5))
acceptance = sns.pointplot(x = 'Column Name', y = 'Percentage of Null Values', data = acceptance_past_applications, marker = '.')
plt.xticks(rotation = 90, fontsize = 5)
acceptance.axhline(10, ls='-', color = 'red')
plt.title('Percentage of Missing Values')
plt.ylabel('Percentage of Null Values')
plt.xlabel('Columns')
plt.show()
```
